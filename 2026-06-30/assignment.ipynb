```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   

    def deposit(self, amount):
        self.__balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient Balance")

    def check_balance(self):
        print("Current Balance:", self.__balance)

account = BankAccount(5000)

account.check_balance()
account.deposit(2000)
account.withdraw(1500)
account.check_balance()
```

```python
import logging
import random

# Configure logging
logging.basicConfig(
    filename="ai_system.log",
    level=logging.INFO,
    format="%(asctime)s - %(message)s"
)

# Spam Detector Class
class SpamDetector:
    def __init__(self, model_name, version):
        self.model_name = model_name
        self.version = version

    def predict(self, message):
        if message.strip() == "":
            raise ValueError("Message cannot be empty.")

        spam_words = ["win", "free", "offer", "prize", "lottery"]

        prediction = "Not Spam"
        for word in spam_words:
            if word in message.lower():
                prediction = "Spam"
                break

        confidence = random.randint(90, 99)

        # Save prediction to file
        file = open("prediction.txt", "a")
        file.write(f"Message: {message}\n")
        file.write(f"Prediction: {prediction}\n\n")
        file.close()

        # Save log
        logging.info(f"Prediction: {prediction} | Confidence: {confidence}%")

        return prediction, confidence


# Function to run the system
def run_detector():
    detector = SpamDetector("Spam Detector", "1.0")

    try:
        message = input("Enter your message: ")

        prediction, confidence = detector.predict(message)

        print("Model:", detector.model_name)
        print("Version:", detector.version)
        print("Prediction:", prediction)
        print("Confidence:", str(confidence) + "%")

    except ValueError as e:
        print("Error:", e)


# Start the program
run_detector()
```

```python
#To keep student information even after the program closes, we use File Handling. The student details are saved to a file (such as students.txt). Since the data is stored on disk, it remains available the next time the program runs.

#Approach
'''
Create a Student class.
Create a student object.
Write the student's details to students.txt.
Read the file whenever you want to view the saved student records.
'''
```

```python
class Student:
    def __init__(self, name, age, course):
        self.name = name
        self.age = age
        self.course = course

    def save(self):
        file = open("students.txt", "a")
        file.write(f"Name: {self.name}, Age: {self.age}, Course: {self.course}\n")
        file.close()


# Create an object
student1 = Student("Rahul", 20, "AI")

# Save student details
student1.save()

# Read and display saved records
file = open("students.txt", "r")
print(file.read())
file.close()
```
