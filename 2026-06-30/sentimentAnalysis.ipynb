```python
#Sentiment Analysis
positive_words = ["good", "great", "excellent", "amazing", "fantastic", "love", "happy", "wonderful", "best", "awesome"]
negative_words = ["bad", "terrible", "awful", "horrible", "worst", "hate", "sad", "disappointing", "poor", "lame"]
email=input("Enter the email text: ").lower()
positive_count=0
negatve_count=0
for word in positive_words:
    if word in email:
        positive_count+=1
for word in negative_words:
    if word in email:
        negatve_count+=1
if positive_count>negatve_count:
    print("The email has a positive sentiment.")
elif negatve_count>positive_count:
    print("The email has a negative sentiment.")
else:
    print("The email has a neutral sentiment.")

    
```

```python
try:
    review = input("Enter a product review: ")
    if review.strip() == "":
        raise ValueError("Review cannot be empty.")
    if review.isdigit():
        raise ValueError("Review cannot be a number.")
    review=review.lower()
    positivewords = ["good", "great", "excellent", "amazing", "fantastic", "love", "happy", "wonderful", "best", "awesome"]
    negativewords = ["bad", "terrible", "awful", "horrible", "worst", "hate", "sad", "disappointing", "poor", "lame"]
    positive_count = sum(word in review for word in positivewords)
    negative_count = sum(word in review for word in negativewords)
    if positive_count > negative_count:
        print("The review has a positive sentiment.")
    elif negative_count > positive_count:
        print("The review has a negative sentiment.")
    else:
        print("The review has a neutral sentiment.")
except ValueError as e:
    print("Input Error:" ,e)
except Exception as e:
    print("An unexpected error occurred:", e)
else:
    print("Sentiment analysis completed successfully.")
finally:
    print("AI Prediction completed.")
    
```
