```python
class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color
    def display(self):
        print("Brand:", self.brand)
        print("Color:", self.color)
        print(" _ " * 30)
car1 = Car("Toyota", "Blue")
car2 = Car("BMW", "Red")
car3=Car("Honda", "Black")
car1.display()
car2.display()
car3.display()
```

```python
#Encapsulation: Wrapping up of data and methods into a single unit.
class AIModel:
    def __init__(self, model_name):
        self.model_name = model_name
    def display(self):
        print("Model Name:", self.model_name)
        print(" _ " * 30)
model=AIModel("Spam Detector")
model.display()
        
```

```python
#Abstraction: Showing only the important details and hiding the unnecessary details.
from abc import ABC, abstractmethod
class AIModel(ABC):
    @abstractmethod
    def predict(self):
        pass
```

```python
#Inheritance: Mechanism of acquiring properties and behaviors from a parent class.
class AIModel:
    def train(self):
        print("Training the AI model...")
class SpamDetector(AIModel):
    pass
spam_detector = SpamDetector()
spam_detector.train()
  
```

```python
#Ploymorphism: Ability of an object to take many forms.
class AIModel:
    def predict(self):
        print("Predicting using the AI model...")
class SpamDetector(AIModel):
    def predict(self):
        print("Predicting spam using the Spam Detector model...")
class SentimentAnalyzer(AIModel):
    def predict(self):
        print("Predicting sentiment using the Sentiment Analyzer model...")
spam=SpamDetector()
sentiment=SentimentAnalyzer()
spam.predict()
sentiment.predict()
```

```python
#str --> Return a user friendly string representation of the object.
#Its used with print(object) or str(object).
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return f"Student Name: {self.name}, Age: {self.age}"
student1 = Student("Alice", 20)
print(student1)
```

```python
num1= int(input("Enter a number: "))
num2= int(input("Enter another number: "))
try:
    #risky code
    print(num1/num2)
except:
    #handling code, execute if an error occurs
    print("An error occurred!")
    
```

```python
try:
    num=int(input("Enter a number: "))
    print(100/num)
except ZeroDivisionError:
    print("Cannot divide by zero!")
except ValueError:
    print("Invalid input! Please enter a valid number.")
else:
    print("Division completed successfully.")
finally:
    print("Execution completed.")
```
