## Question 31

```python
from scipy import stats
import numpy as np

salary = df["Salary"]

standard_error = stats.sem(salary)

print("Standard Error:", standard_error)
```

## Question 32

```python
salary = df["Salary"]

confidence = 0.95

mean = np.mean(salary)
sem = stats.sem(salary)

margin = sem * stats.t.ppf((1 + confidence) / 2, len(salary)-1)

lower = mean - margin
upper = mean + margin

print("95% Confidence Interval:")
print(lower, upper)
```

## Question 33

```python
salary = df["Salary"]

t_stat, p_value = stats.ttest_1samp(salary, 60000)

print("T-Statistic:", t_stat)
print("P-value:", p_value)
```

## Question 34

```python
male = df[df["Gender"]=="Male"]["Salary"]

female = df[df["Gender"]=="Female"]["Salary"]

t_stat, p_value = stats.ttest_ind(male, female)

print("T-Statistic:", t_stat)
print("P-value:", p_value)
```

## Question 37

```python
from sklearn.linear_model import LinearRegression

X = df[["Experience"]]
y = df["Salary"]

model = LinearRegression()

model.fit(X, y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

## Question 38

```python
predicted_salary = model.predict([[8]])

print("Predicted Salary:", predicted_salary[0])
```

## Question 39

```python
r2 = model.score(X, y)

print("R² Score:", r2)
```
