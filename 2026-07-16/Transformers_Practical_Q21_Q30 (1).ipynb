## Question 25

```python
import torch
from transformers import AutoTokenizer, AutoModel
from torch.nn.functional import cosine_similarity

# Load BERT tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

# Sentences with different meanings of "bank"
sents = [
    "I deposited money in the bank.",
    "The fisherman sat near the bank of the river."
]

# Store embeddings
embeds = []

# Extract the embedding for the word "bank"
for s in sents:
    inputs = tokenizer(s, return_tensors="pt")
    outputs = model(**inputs)
    tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])
    idx = tokens.index("bank")
    embeds.append(outputs.last_hidden_state[0, idx])

# Compare the two embeddings
print("Cosine similarity:", cosine_similarity(embeds[0], embeds[1], dim=0))

# Lower similarity shows that BERT creates context-aware embeddings.
```

## Question 28

```python
import torch

# Define input dimensions
batch = 2
seq = 5
embed_dim = 32
heads = 4
head_dim = embed_dim // heads

# Create a random input tensor
X = torch.rand(batch, seq, embed_dim)

# Split input into multiple attention heads
Q = X.view(batch, seq, heads, head_dim).transpose(1, 2)
K = Q
V = Q

# Compute attention scores
scores = torch.matmul(Q, K.transpose(-2, -1)) / head_dim**0.5

# Apply softmax to get attention weights
weights = torch.softmax(scores, dim=-1)

# Compute context vectors
context = torch.matmul(weights, V)

# Concatenate all attention heads
concat = context.transpose(1, 2).contiguous().view(batch, seq, embed_dim)

# Apply the final linear projection
linear = torch.nn.Linear(embed_dim, embed_dim)
output = linear(concat)

# Print output shapes
print("Concatenated:", concat.shape)
print("Projected:", output.shape)
```

## Question 30

```python

import torch
import torch.nn as nn
# Define a GPT-style Decoder Block
class DecoderBlock(nn.Module):

    # Constructor to initialize the layers
    def __init__(self, d_model=32, heads=4):
        super().__init__()

        # Create a Multi-Head Self-Attention layer
        # d_model = embedding dimension
        # heads = number of attention heads
        # batch_first=True expects input as (batch_size, sequence_length, embedding_dimension)
        self.attn = nn.MultiheadAttention(d_model, heads, batch_first=True)

        # Layer Normalization after the attention block
        self.norm1 = nn.LayerNorm(d_model)

        # Feed Forward Neural Network
        self.ff = nn.Sequential(

            # First fully connected layer (32 → 64)
            nn.Linear(d_model, 64),

            # Apply ReLU activation function
            nn.ReLU(),

            # Second fully connected layer (64 → 32)
            nn.Linear(64, d_model)
        )

        # Layer Normalization after the Feed Forward Network
        self.norm2 = nn.LayerNorm(d_model)

    # Forward pass of the decoder block
    def forward(self, x):

        # Get the sequence length from the input tensor
        seq = x.size(1)
        # Create an upper triangular mask for causal (look-ahead) attention
        mask = torch.triu(torch.ones(seq, seq), 1).bool()
        # Apply masked multi-head self-attention
        attn, _ = self.attn(x, x, x, attn_mask=mask)
        # Add residual connection and apply Layer Normalization
        x = self.norm1(x + attn)
        # Pass the result through the Feed Forward Network
        ff = self.ff(x)
        # Add another residual connection and apply Layer Normalization
        x = self.norm2(x + ff)
        return x
model = DecoderBlock()

# Generate a random input tensor
# Shape = (batch_size=2, sequence_length=5, embedding_dimension=32)
inp = torch.rand(2, 5, 32)
# Pass the input through the decoder block
out = model(inp)
print("Input Shape :", inp.shape)
print("Output Shape:", out.shape)
```
