## Download Tokenizer

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

print("Tokenizer Loaded Successfully")
```

## Tokenize One Review

```python
review = dataset["train"][0]["text"]

tokens = tokenizer(review)

print(tokens)
```

## Input IDs

```python
print(tokens["input_ids"][:50])
```

## Token Names

```python
print(tokenizer.convert_ids_to_tokens(tokens["input_ids"][:50]))
```

## Tokenize Entire Dataset

```python
def tokenize(example):

    return tokenizer(
        example["text"],
        truncation=True,
        padding="max_length",
        max_length=128
    )

tokenized_dataset = dataset.map(tokenize, batched=True)
```

## Tokenized Dataset

```python
print(tokenized_dataset)
```
