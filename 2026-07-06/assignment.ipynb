```python
import requests
import pandas as pd
```

```python
url = "https://jsonplaceholder.typicode.com/users"
```

```python
response = requests.get(url)

print("Status Code:", response.status_code)
```

```python
data = response.json()

print(data)
```

```python
df = pd.DataFrame(data)

df.head()
```

```python
df = df[
    [
        "id",
        "name",
        "username",
        "email",
        "phone",
        "website"
    ]
]

df.head()
```

```python
df.to_csv(
    "api_extracted_data.csv",
    index=False
)

print("CSV file saved successfully")
```

```python
# GPS Location Logging
import logging


# Configure logging
logging.basicConfig(
    filename="location_tracking.log",
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S"
)


# Sample GPS Data
vehicle_id = "Vehicle_101"

latitude = 12.9716
longitude = 77.5946

event = "Vehicle location updated"


# Creating log message
message = (
    f"Vehicle ID: {vehicle_id} | "
    f"Event: {event} | "
    f"Latitude: {latitude} | "
    f"Longitude: {longitude}"
)


# Writing log
logging.info(message)


print("Location information logged successfully")
```

```python
# Formatters 

import logging


formatter = logging.Formatter(
    "%(asctime)s | %(levelname)s | %(name)s | %(message)s"
)
```

```python
# Filters
import logging


class ErrorFilter(logging.Filter):

    def filter(self, record):

        return record.levelno >= logging.ERROR



logger = logging.getLogger()


logger.addFilter(
    ErrorFilter()
)


logging.error(
    "Database failure"
)

# Only ERROR and higher-level messages are stored.
```

```python
# Log  rotation
import logging
from logging.handlers import RotatingFileHandler


logger = logging.getLogger(
    "RotationLogger"
)


handler = RotatingFileHandler(
    "application.log",
    maxBytes=1000,
    backupCount=3
)


logger.addHandler(handler)


logger.warning(
    "Application running"
)
```
