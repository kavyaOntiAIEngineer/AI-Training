# Q32. Write simple Python code (using NumPy) to perform one weight update step using the Gradient Descent formula.

```python
import numpy as np

weight = 0.5
gradient = 0.2
learning_rate = 0.1

new_weight = weight - learning_rate * gradient

print("Updated Weight:", new_weight)
```

# Q46. Write Python code using NumPy to implement the Sigmoid function and test it on the value x = 2.

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = 2
print("Sigmoid(", x, ") =", sigmoid(x))
```

# Q47. Write Python code to implement the ReLU function and test it on a list of 5 values (a mix of positive and negative numbers).

```python
import numpy as np

def relu(x):
    return np.maximum(0, x)

values = np.array([-5, -2, 0, 4, 8])

print("Input :", values)
print("ReLU  :", relu(values))
```

# Q48. Why can't a neural network learn complex patterns if it only uses linear activation functions?

```python
If all activation functions are linear, the entire neural network behaves like a single linear model, regardless of the number of layers. Therefore, it cannot learn complex or non-linear relationships in the data.
```
